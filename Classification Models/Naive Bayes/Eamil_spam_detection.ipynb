```python
from nltk.corpus import stopwords
sw = set(stopwords.words("english"))
print(sw)
```

```python
text = "It was a great movie to watch."
```

```python
def remove_stopwords(sents_r,sw_r):
    useful_words_r = []
    for sent in sents_r:
        words = nltk.word_tokenize(sent)
        for word in words:
            if(word not in sw_r and word != "."):
                useful_words_r.append(word)
    return useful_words_r
    
```

```python
useful_words = remove_stopwords(sents,sw)
print(useful_words)
```
